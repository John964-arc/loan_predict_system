```python
#clean
df.columns = df.columns.str.strip()   #gives list of columns with leading and trailing spaces removed
for col in df.select_dtypes(include='object').columns:  
    df[col] = df[col].str.strip()           #removes leading and trailing spaces from all object type columns
```

```python
df.drop(columns=['loan_id'], inplace=True) 
```

```python
df
```

```python
df.isnull().sum()
```

```python
#Encode target variable  ,convert 'Approved' to 1 and 'Rejected' to 0
df['loan_status'] = df['loan_status'].map({
    'Approved': 1, 
    'Rejected': 0
    })  
```

```python
df
```

```python
x = df.drop('loan_status', axis=1)
y = df['loan_status']
df
```

```python
x = pd.get_dummies(x, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42 ,stratify=y)
```

```python
df['loan_status'].value_counts()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
#model training
pipe_lr = Pipeline([          
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(max_iter=1000))
])
```

```python
pipe_lr.fit(x_train, y_train)
```

```python
pipe_rf = Pipeline([
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier())
])
```

```python
pipe_rf.fit(x_train, y_train)
```

```python
pipe_gr = Pipeline([
    
    ('model', GradientBoostingClassifier())
])
```

```python
pipe_gr.fit(x_train, y_train)
```

```python
pipe_lr.named_steps
```

```python
from sklearn.metrics import accuracy_score

models =  {
    "Logistic Regression":pipe_lr,
    "Random forest":pipe_rf,
    "Gradient Boostinf":pipe_gr
}
for name,model in models.items():
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test,y_pred)
    print(f"{name} Accuracy :{acc:4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test,pipe_gr.predict(x_test))

sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("ConfusionMatrixDisplay")
plt.show()
```
